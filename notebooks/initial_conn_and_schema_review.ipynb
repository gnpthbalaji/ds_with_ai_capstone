```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../data/ecommerce.db')

# Check what tables we have
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
print("Tables:", tables)

# Peek at orders
orders = pd.read_sql("SELECT * FROM orders LIMIT 5", conn)
print("\nOrders sample:")
print(orders)

# Check reviews
reviews = pd.read_sql("SELECT review_score, COUNT(*) as count FROM order_reviews GROUP BY review_score", conn)
print("\nReview scores distribution:")
print(reviews)

# Tables and column Info - All Tables
print("\nDatabase Schema:")
# orders
orders_info = pd.read_sql("PRAGMA table_info(orders)", conn)
print("\nOrders table columns:")
print(orders_info)
# customers
customers_info = pd.read_sql("PRAGMA table_info(customers)", conn)
print("\nCustomers table columns:")
print(customers_info)
# order_items
order_items_info = pd.read_sql("PRAGMA table_info(order_items)", conn)
print("\nOrder_items table columns:")
print(order_items_info)
# order_reviews
order_reviews_info = pd.read_sql("PRAGMA table_info(order_reviews)", conn)
print("\nOrder_reviews table columns:")
print(order_reviews_info)
#order_payments
order_payments_info = pd.read_sql("PRAGMA table_info(order_payments)", conn)
print("\nOrder_payments table columns:")
print(order_payments_info)
#products
products_info = pd.read_sql("PRAGMA table_info(products)", conn)
print("\nProducts table columns:")
print(products_info)  
#Geolocation
geolocation_info = pd.read_sql("PRAGMA table_info(geolocation)", conn)
print("\nGeolocation table columns:")
print(geolocation_info)
#Sellers
sellers_info = pd.read_sql("PRAGMA table_info(sellers)", conn)
print("\nSellers table columns:")
print(sellers_info) 
# product category
product_category_info = pd.read_sql("PRAGMA table_info(product_category_name_translation)", conn)
print("\nProduct_category_name_translation table columns:")
print(product_category_info)
conn.close()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../data/ecommerce.db')

# Check what tables we have
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
print("Tables:", tables)

# Peek at orders
orders = pd.read_sql("SELECT * FROM orders LIMIT 5", conn)
print("\nOrders sample:")
print(orders)

# Check reviews
reviews = pd.read_sql("SELECT review_score, COUNT(*) as count FROM order_reviews GROUP BY review_score", conn)
print("\nReview scores distribution:")
print(reviews)

conn.close()
```
